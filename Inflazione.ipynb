I dati dell'inflazione sono disponibili a su questo [sito](https://it.inflation.eu/tassi-di-inflazione/italia/inflazione-storica/cpi-inflazione-italia.aspx).

```python
import pandas as pd
```

```python
# leggere tutte le tabelle
df = pd.read_html("https://it.inflation.eu/tassi-di-inflazione/italia/inflazione-storica/cpi-inflazione-italia.aspx")

# seleziona la tabella della inflazione storica
df = df[8]
```

```python
df
```

```python

df.columns = df.iloc[0]
df = df.reindex(df.index.drop(0)).reset_index(drop=True)
```

```python
# riorganizza colonne
anno = df["inflazione annua (da dic. a dic)"].T.stack().values
inflazione = df["inflazione"].T.stack().values
df = pd.DataFrame({"anno":anno,
                   "inflazione":inflazione})

# convertirsione da stringa a numerico
df["anno"] = df["anno"].str.split(" ").str[-1].astype("int64")
df["inflazione"] = df["inflazione"].str.replace("\xa0%", "").str.replace(",", ".").astype("float64")/100
```

```python
df
```

```python
# seleziona valori dell' inflazione dal 2004 al 2018
vls = df[df["anno"].between(2004,2018)]["inflazione"].values

ci = 160_000
cf = ci
spese = 70_000
mesi = 14 * 12
for v in vls+1:
    cf *= v
diff = cf-ci
tot = spese+diff
print("Capitale iniziale: {:0.2f}".format(ci))
print("Aggiustato inflazione: {:0.2f}".format(cf))
print("Differenza: {:0.2f}".format(diff))
print("Costo totale: {:0.2f}".format(tot))
print("Costo mensuale: {:0.2f}".format(tot/mesi))
```
